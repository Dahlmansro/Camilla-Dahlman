# MNIST Data

In this exercise, some data preprocessing steps are prepared (look it through so you understand it). Thereafter you will create two models. As always, the documentation is the place to go if you need to deepen your understanding or find out how to do something. 

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

from sklearn.svm import LinearSVC
from sklearn.svm import SVC

from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
````

Vi definierar en funktion för att visualisera våra *confusion matrices* senare.

````python
def display_confusion_matrix(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    ConfusionMatrixDisplay(cm).plot()
````

Vi hämtar MNIST-datasetet och kollar dess beskrivning.

````python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, cache=True, as_frame=False)

X = mnist["data"]
y = mnist["target"]
````

````python
from IPython.display import Markdown
display(Markdown(mnist.DESCR))
````

````python
# Taking less data to speed up training. 
X_train = X[:5000]
y_train = y[:5000]
X_val = X[5000:6000]
y_val = y[5000:6000]

X_test = X[6000:7000]
y_test = y[6000:7000]

# Standardizing the data (as you generally always should do when using SVM models).
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)  # Only transforming the validation data. 
X_test_scaled = scaler.transform(X_test)  # Only transforming the test data. 
````

````python
# Plotting the features of the X data, it looks like a 5.
some_digit = X[0]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap=plt.get_cmap("binary"))
# plt.axis("off")
````

````python
#Skriver ut antalet träningsbilder i datasetet samt antalet pixlar per bild (28x28)
print(X_train.shape)
print(y_train.shape)
````

# Creating a simple modell and evaluating it on the validation set

1. Create a linear support vector classifier model (random_state=42 is just used so we get same results due to the training algorithm beeing stochastic.):
```python
lin_clf = LinearSVC(random_state=42)
```

2. Fit the model on the scaled training data. **You might get an warning, ignore it.**
3. Predict the scaled validation data. 
4. Compute accuracy score, use the imported functionality: 
```python
from sklearn.metrics import accuracy_score
```
5. Visualize the confusion matrix, use the function we created: 
```python
def display_confusion_matrix
```

````python
#1. Skapar en linjär SVM-modell
lin_clf = LinearSVC(random_state=42)

#LinearSVC försöker hitta en rak linje (eller hyperplanet) som separerar klasserna.
#Random_state ser till att modellen tränas på exakt samma sätt varje gång vi kör koden
````

````python
#2. Träntar modellen
lin_clf.fit(X_train_scaled, y_train)
````

````python
#3. Förutsäga (predict) den skalade valideringsdatan
y_pred = lin_clf.predict(X_val_scaled)
````

````python
#4. Beräkna accuracy

accuracy_lin = accuracy_score(y_val, y_pred)  # Nu sparas det som accuracy_lin
print("Linear SVC Accuracy:", accuracy_lin)
````

````python
#Visualisera konfusionsmatrisen (för att förstå vilka klasser som blandas ihop)
def display_confusion_matrix(y_true, y_pred):
    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    import matplotlib.pyplot as plt

    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_true), yticklabels=set(y_true))
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.show()

display_confusion_matrix(y_val, y_pred)
````

# Modell Selection with Gridsearch
## Doing GridSearch (Takes a while to run) - remove the comments if you want to run the code

3. Use GridSearchCV and fit the model with the hyperparameter grid above **(note, fitting the model with grid_search takes a while)**.

5. Predict the scaled validation data, compute the accuracy score, visualize the confusion matrix. 


6. Which model is better, the linear created above or this one where we used GridSearch?

````python
#1 Skapa en generell SVC modell (SVC kan använda ickelinjära hyperplan för att separera klasser)
svm_clf = SVC()
````

````python
#2 Specifika hyperparameter-grid som vi vill testa
hyper_param_grid = [
    {'kernel': ['rbf', 'poly'], 'gamma': [1, 2], 'C': [0.5, 1, 1.5]}
]
````

````python
#3 Skapa en GridSearchCV modell

gs_svm = GridSearchCV(SVC(), hyper_param_grid, cv=5, scoring="accuracy", verbose=2)

gs_svm.fit(X_train_scaled, y_train)
````

````python
#4 Analysera resultaten, vilka hyperparametrar ger bäst resultat
print("Bästa parametrar:", gs_svm.best_params_)
print("Bästa score:", gs_svm.best_score_)
````

````python
# 5. Gör prediktioner på valideringsdatan

y_pred_gs = gs_svm.best_estimator_.predict(X_val_scaled)

accuracy_gs = accuracy_score(y_val, y_pred_gs)
print("GridSearch Model Accuracy:", accuracy_gs)

display_confusion_matrix(y_val, y_pred_gs)
````

````python
#6 Vilken modell är bäst?

# Skapa en dictionary med accuracy-resultaten
model_results = {
    "LinearSVC": accuracy_lin,
    "GridSearch SVC (poly)": accuracy_gs
}

# Hitta den bästa modellen
best_model = max(model_results, key=model_results.get)

# Skriv ut resultatet
for model, acc in model_results.items():
    print(f"{model}: {acc:.4f}")

print("\nBästa modellen är:", best_model)
````

# Model Evaluation on Test Data
Use the best model for evaluation on test data. 

````python
#Testar den bästa modellen, GridSearch SVC (poly) med parametrarna → C=0.5, gamma=1, kernel='poly'. på testdatan

y_test_pred = gs_svm.best_estimator_.predict(X_test_scaled)

accuracy_test = accuracy_score(y_test, y_test_pred)
print("Test Accuracy:", accuracy_test)
````

````python
#Visualisera konfusionsmatrisen för testdatan
display_confusion_matrix(y_test, y_test_pred)
````
