### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

#Objektorienterad
x = np.linspace(-4, 4, num=100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Min första plott")
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()

#Pyplot orienterad
x = np.linspace(-4, 4, num=100)
y = x**2

plt.plot(x, y)
plt.title("Min första plott")
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```
